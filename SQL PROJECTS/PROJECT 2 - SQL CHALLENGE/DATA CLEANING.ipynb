# IMPORTING THE DATA

```python
import pandas as pd
df_count = pd.read_csv('countries.csv')
df_city = pd.read_csv('cities.csv')
df_curr = pd.read_csv('currencies.csv')
df_lang = pd.read_csv('languages.csv')
```

# CLEANING THE DATA TABLE-WISE

## COUNTRIES TABLE

```python
df_count['country_name'] = df_count['country_name'].str.lower().replace('[^a-zA-Z ]', '',regex=True)
```

```python
df_count['country_name'] =df_count['country_name'].str.title()
```

```python
df_count['country_name']
```

```python
df_count['country_code_2'] = df_count['country_code_2'].str.lower().replace('[^a-zA-Z ]', '',regex=True)
df_count['country_code_3'] = df_count['country_code_3'].str.lower().replace('[^a-zA-Z ]', '',regex=True)
df_count['region'] = df_count['region'].str.title().replace('[^a-zA-Z ]', '',regex=True)
df_count['sub_region'] = df_count['sub_region'].str.title().replace('[^a-zA-Z ]', '',regex=True)
df_count['intermediate_region'] = df_count['intermediate_region'].str.title().replace('[^a-zA-Z ]', '',regex=True)
df_count['intermediate_region'] = df_count['intermediate_region'].fillna('Not Present')
```

## CITIES TABLE

```python
df_city['insert_date'] = pd.to_datetime(df_city['insert_date'])
df_city['city_name'] = df_city['city_name'].str.title()
df_city['city_name'] = df_city['city_name'].str.title().replace('[^a-zA-Z ]', '',regex=True)
df_city['country_code_2'] = df_city['country_code_2'].str.lower().replace('[^a-zA-Z ]', '',regex=True)
```

## CURRENCIES TABLE

```python
df_curr['country_code_2'] = df_curr['country_code_2'].str.lower().replace('[^a-zA-Z ]', '',regex=True)
df_curr['currency_code'] = df_curr['currency_code'].str.lower().replace('[^a-zA-Z ]', '',regex=True)
```

```python
df_curr[df_curr['currency_name'] == 'us dollar']
df_curr['currency_name'] = df_curr['currency_name'].str.title().replace('[^a-zA-Z ]', '',regex=True)
df_curr['currency_name']
df_curr['currency_name'] = df_curr['currency_name'].replace('Us Dollar', 'US Dollar')
```

## LANGUAGES TABLE

```python
df_lang.head()
df_lang['language'] = df_lang['language'].str.title().replace('[^a-zA-Z ]', '',regex=True)
df_lang['country_code_2'] = df_lang['country_code_2'].str.lower().replace('[^a-zA-Z ]', '',regex=True)
```

# HERE ARE DATA IS CLEANED, NOW WE HAVE TWO CHOICES:-
- TO TRANSFER IT TO CSV FILES AND THEN IMPORT IT THROUGH AN IMPORT WIZARD IN YOUR SQL DATABASE
- TO ESTABLISH A CONNECTION AND SEND DATA DIRECTLY TO OUR SQL DATABASE

```python
df_count.to_csv("COUNTRIES.csv", index = False)
df_curr.to_csv("CURRENCIES.csv", index = False)
df_lang.to_csv("LANGUAGES.csv", index = False)
df_city.to_csv("CITIES.csv", index = False)
```

### CREATING THE CONNECTION STRING

```python
from sqlalchemy import create_engine
engine = create_engine("mssql+pyodbc://@localhost\\SQLEXPRESS/github_practice?driver=ODBC+Driver+18+for+SQL+Server&trusted_connection=yes&TrustServerCertificate=yes")
```

```python
df_count.to_sql(name = "countries", con=engine, index = False, if_exists = 'replace')
```

```python
df_curr.to_sql(name = "currencies", con=engine, index = False, if_exists = 'replace')
df_lang.to_sql(name = "languages", con=engine, index = False, if_exists = 'replace')
df_city.to_sql(name = "cities", con=engine, index = False, if_exists = 'replace')
```
